```python
# Average Career Length
df["career_length"].mean()
```

```python
# Highest Inns Score by country
df.groupby("Country")["Highest_Inns_Score"].max().sort_values(ascending=False)
```

```python
# Hundreds, Fifties, ducks(0) AVG by country
df.groupby("Country")[["100", "50", "0"]].mean(numeric_only=True)
```
