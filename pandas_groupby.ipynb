```python
dftransform = sns.load_dataset("tips")
```

```python
dftransform.head()
```

```python
dftransform["medianbill"] = df.groupby("time", observed=True)["total_bill"].transform("median")
```

```python
dftransform.head()
```

```python
dftransform.query("time == 'Lunch'")
```

```python
dftransform.query("time == 'Dinner'")
```

```python
dftransform.query("total_bill > medianbill")
```

```python
# same as above code
dftransform[ dftransform["total_bill"] > dftransform["medianbill"] ]
```

```python
dftransform["zscore"] = df.groupby("time", observed=True)["total_bill"].transform(lambda x: (x - x.mean()) / x.std())
```

```python
dftransform
```

```python
dftransform.groupby("time", observed=True).apply(lambda x: (x["tip"] / x["total_bill"]).mean()).to_frame("avgtippercent").reset_index()
```

```python
dftransform.groupby("time", observed=True).apply(lambda x: (100 * x["tip"] / x["total_bill"]).mean()).to_frame("avgtippercent").reset_index()
```
